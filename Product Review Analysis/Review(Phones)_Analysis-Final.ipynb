```python
print('Number of Phone =>', len(reviews_df['Phone Name'].unique()))
# converting phone name to lower case
reviews_df['Phone Name'] = reviews_df['Phone Name'].str.lower()

def clean_text(row):
    # removing non ascii characters 
    text = row.encode('ascii', 'ignore')

    # decoding encoded characters
    text = text.decode('ascii')
    
    # removing new line and tabs
    text = text.replace('\n', ' ').replace('\r', '').replace('\t', ' ')
    
    return text

def count_word(row):
    # count words in reviews
    count = len(row.split(' '))
    
    return count

reviews_df['Review'] = reviews_df['Review'].apply(clean_text)
reviews_df['Words Count'] = reviews_df['Review'].apply(count_word)

# Removing short reviews as we need to summarize review to around 100-120 words
reviews_df = reviews_df[reviews_df['Words Count'] > 100]
reviews_df = reviews_df[reviews_df['Words Count'] < 10000] 
reviews_df = reviews_df[reviews_df['Phone Name'] != 'android 12'] 
```

```python
import pandas as pd
import re
import torch
from transformers import BartTokenizer, BartForConditionalGeneration, BartConfig
import numpy as np

model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')
tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')
```

```python
available_phones = sorted(reviews_df['Phone Name'].unique())
def text_summarization_Bart(user_input):
    if user_input.lower() in available_phones:
        user_input = user_input.lower()
        
        filtered_df = reviews_df[reviews_df['Phone Name']==user_input]
        articles_to_summarize = list(filtered_df['Review'])
        
        article_id = []
        total_input = np.array([])
        
        for a_id, article in enumerate(articles_to_summarize):
            inputs = tokenizer.batch_encode_plus([article], max_length=1024, return_tensors='pt', return_overflowing_tokens=True, pad_to_max_length=True)
            article_id.append(a_id)
            
            overflowing_tokens = np.squeeze(np.array(inputs['overflowing_tokens']))
            trunc_input = []
            for i in range(int(inputs['num_truncated_tokens'][0])//1024):
                trunc_input.append(overflowing_tokens[i*1024:(i+1)*1024])
                article_id.append(a_id)
                
            remaining_inputs = tokenizer.batch_encode_plus([tokenizer.decode(overflowing_tokens[(i+1)*1024:], skip_special_tokens=True)], max_length=1024, return_tensors='pt', pad_to_max_length=True)
            if int(remaining_inputs['input_ids'].shape[1])>512:
                trunc_input.append(np.squeeze(np.array(remaining_inputs['input_ids'])))
                article_id.append(a_id)
            
            trunc_input = np.array(trunc_input)
            article_input = np.vstack((np.array(inputs['input_ids']), trunc_input))
            
            if a_id == 0:
                total_input = article_input.copy()
            else:
                total_input = np.vstack((total_input,article_input))
                
        total_input = torch.tensor(total_input)
        
        # Summarization
        summary_ids = model.generate(total_input, num_beams=4, max_length=150, early_stopping=True)
        summary_array = [tokenizer.decode(g, skip_special_tokens=True, clean_up_tokenization_spaces=False) for g in summary_ids]

        article_id = np.array(article_id)
        summary_array = np.array(summary_array)

        summary_dict = {"Source":[],"URL":[],"Summary":[]}

        for i in set(article_id):
            cor_articles = summary_array[article_id==i]
            summary = " ".join(list(cor_articles))
            summary_dict['Source'].append(list(filtered_df['Source'])[i])
            summary_dict['URL'].append(list(filtered_df['URL'])[i])
            summary_dict['Summary'].append(summary)

        summary_df = pd.DataFrame(summary_dict)
        return summary_df
            
    else:
        raise Exception("We cuurently don't have any reviews for device %s in our database."%user_input)
```

```python
# google pixel 6 example for text summarization 
summary_df = text_summarization_Bart('google pixel 6')
summary_df.to_csv("output/google_pixel_6.csv")
summary_df
```

### TextRank Algorithm

```python
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
import networkx as nx

def text_summarization_TextRank(text):
    sentences = []
    sentences.append(sent_tokenize(text)) 
    
    # flatten the list
    sentences = [item for sublist in sentences for item in sublist]

    # remove punctuations, numbers and special characters
    sentences_clean = pd.Series(sentences).str.replace("[^a-zA-Z]", " ")

    # make alphabets lowercase
    sentences_clean = [s.lower() for s in sentences_clean]

    # removing stop words
    sentences_clean = [remove_stopwords(r.split()) for r in sentences_clean]

    sentence_vectors = []
    for i in sentences_clean:
        if len(i) != 0:
            v = sum([word_embeddings.get(w, np.zeros((100,))) for w in i.split()])/(len(i.split())+0.001)
        else:
            v = np.zeros((100,))
        sentence_vectors.append(v)
    
    # calculating similarity matrix
    sim_mat = np.zeros([len(sentences), len(sentences)])
    from sklearn.metrics.pairwise import cosine_similarity
    for i in range(len(sentences)):
        for j in range(len(sentences)):
            if i != j:
                sim_mat[i][j] = cosine_similarity(sentence_vectors[i].reshape(1,100), sentence_vectors[j].reshape(1,100))[0,0] 

    nx_graph = nx.from_numpy_array(sim_mat)
    scores = nx.pagerank(nx_graph)
    top_ranked_sentences = sorted(((scores[i],s) for i,s in enumerate(sentences)), reverse=True)
    
    text_summary = ''
    for i in range(4):
        text_summary += top_ranked_sentences[i][1]
    
    return text_summary

def remove_stopwords(sen):
    sen_new = " ".join([i for i in sen if i not in stop_words])
    return sen_new
```

```python
def text_summarization_all_TextRank(phone):
    summary_dict = {"Phone Name":[],"Summary":[]}
    temp_df = df[df['Phone Name'] == phone ]
    review = ''
    for ph in temp_df.itertuples():
        review += ph.Review
    textrank_summary = text_summarization_TextRank(review)
    summary_dict["Phone Name"].append(phone)
    summary_dict["Summary"].append(textrank_summary)
    summary_df = pd.DataFrame(summary_dict)
    return summary_df

summary_df_extractive = text_summarization_all_TextRank('google pixel 6')
summary_df_extractive.to_csv("output/extractive_google_pixel6.csv")
summary_df_extractive
```

# Text Sentiment Analysis 

### VADER

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import tokenize
import copy

with open("negative-words.txt",'r') as f:
    features=[line.strip() for line in f]
    
compititor_phone = sorted(reviews_df['Phone Name'].unique())

def get_list(p_df, compititor_phone):
    positive_feature = []
    compititor_inText = []
    for row in p_df.itertuples():
        for feature in features:
            if feature in row.Sentence.lower():
                positive_feature.append(feature)
        for compititor in compititor_phone:
            if compititor in row.Sentence.lower():
                compititor_inText.append(compititor)
            
    return positive_feature, compititor_inText


def sentiment_analysis_VADER(text, phone):
    lines_list = tokenize.sent_tokenize(text)
    # initalize analyzer
    sid = SentimentIntensityAnalyzer()
    senti_df = []
    # analyze the sentiment sentence by sentence
    for sentence in lines_list:
        senti_list = [sentence]
        st = sid.polarity_scores(sentence)
        for k in sorted(st):
            senti_list.append(st[k]) 
        senti_df.append(senti_list)

    pn_df = pd.DataFrame(senti_df, columns = ['Sentence', 'Compound','Negative','Neutral','Positive'])
    c_phone = copy.deepcopy(compititor_phone)
    c_phone.remove(phone)
    pdf = pn_df[pn_df['Compound'] > 0]
    pdf = pdf.sort_values(by=['Positive'], ascending=False)
    ndf = pn_df[pn_df['Compound'] < 0]
    pdf = pdf.sort_values(by=['Compound'])
    p_list, p_c_list = get_list(pdf, c_phone)
    n_list, n_c_list = get_list(ndf, c_phone)
    
    return p_list, p_c_list, n_list, n_c_list, pdf.head(5), ndf.head(5)
```

### TextBlob

```python
from textblob import TextBlob
from textblob import Blobber
from textblob.sentiments import NaiveBayesAnalyzer

def sentiment_analysis_TextBlob(text, ph):
    lines_list = tokenize.sent_tokenize(text)

    senti_df = []
    # initalize analyzer
    blob = Blobber(analyzer=NaiveBayesAnalyzer())
    for sentence in lines_list:
        ss =  blob(sentence).sentiment
        # analyze the sentiment sentence by sentence
        senti_df.append([sentence, ss[0], ss[1]])

    pn_df = pd.DataFrame(senti_df, columns = ['Sentence', 'Polarity', 'Subjectivity'])
    c_phone = copy.deepcopy(compititor_phone)
    c_phone.remove(ph)
    
    pdf = pn_df[pn_df['Polarity'] == 'pos']
    pdf = pdf.sort_values(by=['Subjectivity'], ascending=False)
    ndf = pn_df[pn_df['Polarity'] == 'neg']
    ndf = ndf.sort_values(by=['Subjectivity'], ascending=False)
    
    p_list, p_c_list = get_list(pdf, c_phone)
    n_list, n_c_list = get_list(ndf, c_phone)
    
    return p_list, p_c_list, n_list, n_c_list, pdf.head(5), ndf.head(5)
        
```

```python
import pandas
from collections import Counter
from itertools import islice

# For plotting graph of list of feature and compititor
def plot_graph(p_list, n_list, p_c_list, n_c_list):
    
    p_list = Counter(p_list)
    p_list = dict(islice(p_list.most_common(), 5))  if len(p_list.most_common()) > 5 else dict(p_list.most_common())
    
    n_list = Counter(n_list)
    n_list = dict(islice(n_list.most_common(), 5))  if len(n_list.most_common()) > 5 else dict(n_list.most_common())

    p_c_list = Counter(p_c_list)
    p_c_list = dict(islice(p_c_list.most_common(), 5))  if len(p_c_list.most_common()) > 5 else dict(p_c_list.most_common())

    n_c_list = Counter(n_c_list)
    n_c_list = dict(islice(n_c_list.most_common(), 5))  if len(n_c_list.most_common()) > 5 else dict(n_c_list.most_common())

    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 20))

    df_g = pandas.DataFrame.from_dict({'features':p_list.keys(), 'features count':p_list.values()})
    df_g.plot.bar(x='features', y='features count', rot=30, color='green', ax=axes[0, 0], title='Showing feature mentioned with positive sentences.')

    df_g = pandas.DataFrame.from_dict({'features':n_list.keys(), 'features count':n_list.values()})
    df_g.plot.bar(x='features', y='features count', rot=30, color='red', ax=axes[0, 1],  title='Showing feature mentioned with negative sentences.')

    df_g = pandas.DataFrame.from_dict({'competitor':p_c_list.keys(), 'competitor count': p_c_list.values()})
    df_g.plot.bar(x='competitor', y='competitor count', rot=30, color='green', ax=axes[1, 0],  title='Showing compititor mentioned with positive sentences.')

    df_g = pandas.DataFrame.from_dict({'competitor':n_c_list.keys(), 'competitor count': n_c_list.values()})
    df_g.plot.bar(x='competitor', y='competitor count', rot=30, color='red', ax=axes[1, 1], title='Showing compititor mentioned with negative sentences.')
```

#### Outcome of sentiment analysis

```python
# Performing sentiment analysis on all phone useing VADER
def calculate_sentiment_all_VADER():
    summary_dict = {"Phone Name":[],"Positive_Sentences":[], "Negative_Sentences":[], 'Feature_Positive': [], 'Feature_Negative': [], 'Compititor_Positive': [], 'Compititor_Negative': []}
    for phone in available_phones:
        temp_df = df[df['Phone Name'] == phone ]
        reviews = ''
        for ph in temp_df.itertuples():
            reviews += ph.Review
        p_list, p_c_list, n_list, n_c_list, top_pos, top_neg =  sentiment_analysis_VADER(reviews, phone)
        
        summary_dict["Phone Name"].append(phone)
        summary_dict["Positive_Sentences"].append(top_pos['Sentence'].tolist())
        summary_dict["Negative_Sentences"].append(top_neg['Sentence'].tolist())
        summary_dict["Feature_Positive"].append(p_list)
        summary_dict["Feature_Negative"].append(n_list)
        summary_dict["Compititor_Positive"].append(p_c_list)
        summary_dict["Compititor_Negative"].append(n_c_list)

    summary_df = pd.DataFrame(summary_dict)
    summary_df.to_csv("output/sentiment_df_VADER.csv")
    return summary_df

sentiment_df_VADER = calculate_sentiment_all_VADER()
sentiment_df_VADER
```

```python
# Performing sentiment analysis on all phone useing TextBlog
def calculate_sentiment_all_TextBlob():
    summary_dict = {"Phone Name":[],"Positive_Sentences":[], "Negative_Sentences":[], 'Feature_Positive': [], 'Feature_Negative': [], 'Compititor_Positive': [], 'Compititor_Negative': []}
    for phone in available_phones:
        temp_df = df[df['Phone Name'] == phone ]
        reviews = ''
        for ph in temp_df.itertuples():
            reviews += ph.Review
        p_list, p_c_list, n_list, n_c_list, top_pos, top_neg =  sentiment_analysis_TextBlob(reviews, phone)
        
        summary_dict["Phone Name"].append(phone)
        summary_dict["Positive_Sentences"].append(top_pos['Sentence'].tolist())
        summary_dict["Negative_Sentences"].append(top_neg['Sentence'].tolist())
        summary_dict["Feature_Positive"].append(p_list)
        summary_dict["Feature_Negative"].append(n_list)
        summary_dict["Compititor_Positive"].append(p_c_list)
        summary_dict["Compititor_Negative"].append(n_c_list)

    summary_df = pd.DataFrame(summary_dict)
    summary_df.to_csv("output/sentiment_df_TextBlob.csv")
    return summary_df

sentiment_df_TextBlob = calculate_sentiment_all_TextBlob()
sentiment_df_TextBlob
```
